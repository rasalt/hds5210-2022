This section obtains population data for every state from a public website.
There is only 1 table on this website with no mention of a name so it became intuitive to access the only table at index 0

In the larger scheme of things, this population data is obtained to draw a relative percentage per state.
Reference link: http://www.stats.indiana.edu/population/poptotals/historic_counts_states.asp

```python
!pip3 install lxml
!pip3 install html5lib
!pip3 install beautifulsoup4
```

```python
import pandas as pd
dfs = pd.read_html('http://www.stats.indiana.edu/population/poptotals/historic_counts_states.asp')
poptable = dfs[0]
poptable.info()
poptable.head()
```

Let's keep this data in this table
1. Only state data (not country)
2. Only for years 2010 and 2020

```python
poptable = poptable[['State','2010', '2020']]
poptable = poptable[poptable['State']!= 'United States']
poptable
```
