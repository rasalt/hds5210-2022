````python
def getsexscore(sex):
    """
    Returns the score for the sex
    Parameters:
        sex (string): A string indicating Male of Female
    >>> getagescore("MaLe")
    1
    >>> getagescore("female")
    0  
    """
    try:
#        print("Sex is {}".format(sex))
        if sex.lower() == "male":
            score = 1
        elif sex.lower() == "female":
            score = 0
        else:
            raise ValueError("The value of " + sex + " for the sex variable. It needs to be male or female.")  
#        print("Score is {}".format(score))
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))

def getagescore(age):
    """
    Returns the score for the age
    Parameters:
        age (integer): An age represented in number of years
    >>> getagescore(20)
    0
    >>> getagescore(83)
    4          
    """
#    print("Age is {}".format(age))
    try:
        age = int(age)
        if (age < 16):
            raise ValueError("The value of age is " + age + ". Valid values are >=16")
        if (age >=16) and (age<=49):
            score = 0
        if (age >=50) and (age<=65):
            score = 2
        if (age >=66) and (age<=80):
            score = 3
        if (age >80):
            score = 4            
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))

def getrrpmscore(rrpm):
    """
    Returns the score for the rrpm
    Parameters:
        rrpm (integer): Respiratory rate in breaths per minute
    >>> getrrpmscore(20)
    0
    >>> getrrpmscore(83)
    4          
    """

    try:
        rrpm = int(rrpm)
        
#        print("rrpm is {}".format(rrpm))
        if (rrpm < 0):
            raise ValueError("The rrpm input " + str(rrpm) + " is negative")
        elif (rrpm < 9):
            score = 3
        elif (rrpm >=9) and (rrpm<=11):
            score = 1
        elif (rrpm >=12) and (rrpm<=20):
            score = 0
        elif (rrpm >=21) and (rrpm<=24):
            score = 2
        elif (rrpm >24):
            score = 3   
        else:
            raise Exception("The input of rrpm  is " + rrpm + ".Something is wrong")
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))

def getosscore(os):
    """
    Returns the score for the oxygen saturation
    Parameters:
        os (integer): Respiratory rate in breaths per minute
    >>> getosscore(96)
    0
    >>> getosscore(91)
    3          
    """

    try:
        os = float(os)
        
#        print("o2 sat is {}".format(os))
        if (os < 0):
            raise ValueError("The os input " + str(os) + " is negative")
        elif (os >1):
            raise ValueError("The os input " + str(os) + " is > 100%")
        elif (os > 0.95):
            score = 0
        elif (os >= 0.94) and (os<=0.95):
            score = 1
        elif (os >= 0.92) and (os<=0.93):
            score = 2
        elif (os < 0.92):
            score = 3          
        else:
            raise Exception("The input of os  is " + os + ".Something is wrong")
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))

def gethrbpmscore(hrbpm):

    """
    Returns the score for the heart rate bpm
    Parameters:
        hrbpm (integer): Heart rate, beats per min
    >>> gethrbpmscore(40)
    3
    >>> gethrbpmscore(135)
    3 
    """

    try:
        hrbpm = int(hrbpm)
        
#        print(hrbpm)
        if (hrbpm < 0):
            raise ValueError("The hrbpm input " + str(hrbpm) + " is negative")
        elif (hrbpm < 41):
            score = 3
        elif (hrbpm >= 41) and (hrbpm<=50):
            score = 1
        elif (hrbpm >= 51) and (hrbpm<=90):
            score = 0
        elif (hrbpm >= 91) and (hrbpm<=110):
            score = 1
        elif (hrbpm >= 111) and (hrbpm<=130):
            score = 2
        elif (hrbpm > 130):
            score = 3            
        else:
            raise Exception("The input of hrbpm  is " + hrbpm + ".Something is wrong")
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))
        
def getsysscore(sys):

    """
    Returns the score for the heart rate bpm
    Parameters:
    sys (integer): Systolic BP, mmHg
    >>> gethrbpmscore(96)
    0
    >>> gethrbpmscore(91)
    3          
    """

    try:
        sys = int(sys)
        
#        print(sys)
        if (sys < 0):
            raise ValueError("The sys input " + str(hrbpm) + " is negative")
        elif (sys < 91):
            score = 3
        elif (sys >= 91) and (sys<=100):
            score = 2
        elif (sys >= 101) and (sys<=110):
            score = 1
        elif (sys >= 111) and (sys<=219):
            score = 0
        elif (sys > 219):
            score = 3          
        else:
            raise Exception("The input of sys  is " + sys + ".Something is wrong")
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))     

def gettempscore(temp):

    """
    Returns the score for the temperature
    Parameters:
        temp (float): Temperature - degrees 
    >>> gettempcore(38.4)
    1
    >>> gettempcore(36)
    1
    """

    try:
        temp = float(temp)
        
#        print(temp)
        if (temp < 35.1):
            score = 3
        elif (temp >= 35.1) and (temp<=36.0):
            score = 1
        elif (temp >= 36.1) and (temp<=38.0):
            score = 0
        elif (temp >= 38.1) and (temp<=39.0):
            score = 1
        elif (temp > 39):
            score = 2          
        else:
            raise Exception("The input of temperature  is " + temp + ".Something is wrong")
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))     

        
def getalertdescscore(alertdesc):
    """
    Returns the score for the alertness 
    Parameters:
        alertdesc (string): Alertness 	Alert 	0, Confused or not alert 	3
    >>> getalertdescscore("Confused")
    3
    >>> getalertdescscore("Alert")
    0
    """
    try:
#        print(alertdesc)
        if (alertdesc.lower() == "alert"):
            score = 0
        elif (alertdesc.lower() in ["confused","not alert","confused or not alert"]):
            score = 3        
        else:
            raise Exception("The input of alert description is " + alertdesc + ".Something is wrong")
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))     

def getispo2score(ispo2):
    """
    Returns the score for the inspired oxygen  
    Parameters:
        ispo2 (string): Air 	0, Supplemental oxygen 	2
    >>> getispo2score("AIR")
    0
    >>> getispo2score("supplemental oxygen")
    2
    """
    try:
#        print(ispo2)
        if (ispo2.lower() == "air"):
            score = 0
        elif (ispo2.lower() == "supplemental oxygen"):
            score = 2
        else:
            raise Exception("The input of inspired oxygen is " + ispo2 + ".Something is wrong")
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))   

        
def getperfscore(perf):
    """
    Returns the score for the performance  
    Parameters:
        perf (string):  Unrestricted normal activity 	0
                            Limited strenuous activity, can do light activity 	1
                            Limited activity, can self-care 	2
                            Limited self-care 	3
                            Bed/chair bound, no self-care 	4
    >>> getperfscore("Unrestricted normal activity ")
    0
    >>> getperfscore("Limited activity, can self-care")
    2
    """
    try:
#        print(perf)
        if (perf.lower() == "unrestricted normal activity"):
            score = 0
        elif (perf.lower() == "limited strenuous activity, can do light activity"):
            score = 1
        elif (perf.lower() == "limited activity, can self-care"):
            score = 2
        elif (perf.lower() == "limited self-care"):
            score = 3
        elif (perf.lower() == "bed/chair bound, no self-care"):
            score = 4            
        else:
            raise Exception("The input of performance score is " + perf + ".Something is wrong")
        return score

    except Exception as e:        
        print("Check your inputs")
        print(repr(e))      
````

## Step 1: Calculate PRIEST Clinical Severity Score

This scoring algorithm can be found [here on the MDCalc website](https://www.mdcalc.com/priest-covid-19-clinical-severity-score#evidence).  

1. You will need to write a function called **priest()** with the following input parameters.

 * If score is between 17 and 25, percentile should be 0.59

````python
if 0:
    print("Sex ocre is {}".format(getsexscore("fesfs")))
    print("Age Socre is {}".format(getagescore(84)))
    print("RRPM Socre is {}".format(getrrpmscore("28")))
    print("OS Socre is {}".format(getosscore("28")))
    print("HR BPM Socre is {}".format(gethrbpmscore(100)))
    print("Systolic Socre is {}".format(getsysscore(220)))
    print("Temperature Socre is {}".format(gettempscore(37)))
    print("Alert description Score is {}".format(getalertdescscore("alert")))
    print("Inspired o2 description Score is {}".format(getispo2score("AIR")))
    print("Perf score {}".format(getperfscore("limited self-care")))

score_lookup = {
    0: 1,
    1: 1,
    2: 2,
    3: 2,
    4: 3,
    5: 9,
    6: 15,
    7: 18,
    8: 22,
    9: 26,
    10: 29,
    11: 34,
    12: 38,
    13: 46,
    14: 47,
    15: 49,
    16: 55,
    17: 59,
    18: 59,
    19: 59,
    20: 59,
    21: 59,
    22: 59,
    23: 59,
    24: 59,
    25: 59
}

def priest(sex, age, rrpm, os, hrbpm, sys, temp, alertdesc, ispo2, perf):
    """
    Returns the sum of two decimal numbers in binary digits.
    Parameters:
            sex (string): A string indicating Male of Female
            age (int): Age 
            rrpm (int): Respiratory rate in breaths per minute
            os (float): Oxygen saturation as a percent between 0 and 1
            hrbpm(int): Heart rate in beats per minute
            sys(int): Systolic BP in mmHg
            temp(float): Temperature in degrees C
            alertdesc(string): Alertness as a string description
            ispo2(strin): Inspired Oxygen as as string description
            perf(string): Performance Status as a string description

    Returns:
        A PRIEST score as per this link (https://www.mdcalc.com/priest-covid-19-clinical-severity-score#evidence)
   
    >>> priest("female", 52, 24, .96, 96, 120, 36.5, "Alert", "AIR", "Bed/chair bound, no self-care")
    0.26
    >>> priest("male", 52, 24, .96, 110, 120, 36.5, "Alert", "AIR", "Bed/chair bound, no self-care")
    0.29
    >>> priest("male", 95, 120, .90, 230, 230, 60, "confused", "Supplemental oxygen", "Bed/chair bound, no self-care")
    0.99
    """
    # Check th type of arguments and throw an exception
    try:
        totalscore = getsexscore(sex) + \
                    getagescore(age) + \
                    getrrpmscore(rrpm) + \
                    getosscore(os) + \
                    gethrbpmscore(hrbpm) + \
                    getsysscore(sys) + \
                    gettempscore(temp) + \
                    getalertdescscore(alertdesc) + \
                    getispo2score(ispo2) + \
                    getperfscore(perf) 
        
        #print("PRIEST TOTAL is {}".format(totalscore))
        # Do the lookup to map this total to a percentage
        assert (type(totalscore) is int), "Totalscore should be a valid integer"
        if (totalscore < 26):
            risk = score_lookup[totalscore]
        elif (totalscore > 25):
            risk = 99

        #return totalscore, risk/100.0   
        return risk/100.0
    except Exception as e:        
        print("priest: Erroneous inputs")
        print(repr(e))  


        
#pscore, risk = priest("female", 52, 24, 96, 96, 120, 36.5, "Alert", "AIR", "Bed/chair bound, no self-care")
#print("PRIEST Score is {} and pct risk is {}".format(pscore, risk))

#risk = priest("male", 95, 120, .90, 230, 230, 60, "confused", "Supplemental oxygen", "Bed/chair bound, no self-care")
#print("Priest risk pct is {}".format(risk))
#print("--------------------")
````

````python
def find_hospital(age, sex, risk):
    """
    Returns the hospital matched to a patient's personal information and risk score 

    Parameters:
        age (integer):  Age in years
        sex (string): A string indicating Male of Female
        risk (float): PRIEST risk score (< 1) 
           
    >>> find_hospital(90, "Female", 0.15)
    'Wesley Woods Geriatric Hospital'
    >>> find_hospital(90, "male", 0.99)
    'Wesley Woods Geriatric Hospital'
    >>> find_hospital(23, "male", 0.18)
    'Southwest Hospital and Medical Center'
    >>> find_hospital(21, "female", 0.99)
    'Emory Dunwoody Medical Center'
    >>> find_hospital(45, "female", 0.99)
    'Emory Dunwoody Medical Center'
    """
    
    URL = "https://oumdj6oci2.execute-api.us-east-1.amazonaws.com/prd/?"
    # Check a few inputs
    assert (risk < 1), "Risk should be a number between 0 and 1"
    assert (sex.lower() in  ["male","female"]), "Sex should be male or female"
    
    # Construct the url
    url = URL + "age=" + str(age) + "&sex="+ sex + "&risk_pct="+str(risk)
    #print("URL is {}".format(url))

    try:       
        import requests
        import json
    
        response = requests.get(url)
        data = response.json()
        if data["hospital"]:
            return data["hospital"].strip()

    except Exception as e:        
        print("find_hospital: Arguments {}, {}, {}. Check input ".format(age, sex,risk))
        print(repr(e))   

    
#find_hospital(90, "Female", 0.15)
#find_hospital(90, "male", 0.99)
#find_hospital(23, "male", 0.18)
#find_hospital(21, "female", 0.99)
#find_hospital(45, "female", 0.99)



    
````

## Part 3: Get the address for that hospital from a webpage

Great! Now we have code to tell us which hospital to send someone to... but we don't know where that hospital is. The next function we need to create is one that looks up the address of that hospital.  All of these hospitals are in Atlanta, Georgia.  We're going to use the list on this webpage to lookup the address for that hospital, based on its name.

https://www.officialusa.com/stateguides/health/hospitals/georgia.html

1. You need to create a function called **get_address()** that takes hospital name as a parameter and searches the data on the webpage above to find the addres for that hospital.
2. I've said that all the hospitals are in Atlanta, but this webpage has hospitals from all of Georgia.  So, make sure you verify that the row of data you're using is in Atlanta, just in case there are hospitals with the same name in different cities.
3. Your code will have to parse the HTML on the webpage and turn that into some kind of data structure that you can search through to find the right hospital.
4. If you do find more than one matching hospital in Atlanta with the same name, you should raise a KeyError.
5. If the hospital name isn't found, the function should raise a KeyError.
6. Be sure to use good docstring documentation and includes at least 3 test cases.

````python
def get_address(hospital):
    """
    Returns the address of the hospital name (input) in Atlanta 

    Parameters:
        hospital (string):  Hospital name
           
    >>> get_address("Wesley Woods Geriatric Hospital")
    '1821 CLIFTON ROAD, NE'
    >>> get_address("Hughes Spalding Childrens Hospital")
    '1711 TULLIE CIRCLE NE'
    >>> get_address("Emory Dunwoody Medical Center")
    '4500 NORTH SHALLOWFORD ROAD'
    >>> get_address("NON EXISTENT")
    get_address: Arguments NON EXISTENT
    KeyError('No hospitals in Atlanta with that name',)
    """

    import pandas as pd
    import requests
    URL = "https://www.officialusa.com/stateguides/health/hospitals/georgia.html"
    # Since this page is not straight forard to read, I will read this as a browser text
    header = {
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
        "X-Requested-With": "XMLHttpRequest"
    }
    try:

        r = requests.get(URL, headers=header)
        #URL = "https://en.wikipedia.org/wiki/Texas"
        dfs = pd.read_html(r.text)
        h_table = dfs[0]
    #    h_table.info()
        rslt_df = h_table[(h_table['Hospital Name'] == hospital.upper()) &
                          (h_table['City'] == "Atlanta".upper())]

        numrows = rslt_df.shape[0]        
        if (numrows == 0):
            raise KeyError("No hospitals in Atlanta with that name")
        elif (numrows > 1):
            raise KeyError("More than 1 hospital with that name in Atlanta")
        address = rslt_df['ADDRESS'].to_string(index=False)    
        return address.strip()
    except Exception as e:
        print("get_address: Arguments {}".format(hospital))
        print(repr(e))   

#print(get_address("Wesley Woods Geriatric Hospital"))
#print(get_address("Hughes Spalding Childrens Hospital"))
#print(get_address("fdgfgd"))
````

````python
def process_people(file):
    """
    Uses the input filename, iterates over each row item and creates an output structure with additional infomation
    i.e. with the risk score, hospital and address.
    Returns a dictionary with the patient id as a key and the value being the original patient information along with additional content of 
       risk score, hospital name and the hospital address.


    Parameters:
        file (file):  | delimited file with patient information per row
    """
    import csv
    import json
    results = {}
    with open(file) as f:
        reader = csv.reader(f, delimiter='|')
        header = next(reader)
    #    print("Header is \n --  {}".format(header))       
        for line in reader:
      #      print(line)
            pid, sex, age, breath, o2sat, heart, systolic, temp, alertness, inspired, status = line[0], line[1], line[2], line[3], \
                    line[4],line[5],line[6],line[7], line[8], line[9], line[10]
      #      print(pid, sex, age, breath, o2sat, heart, systolic, temp, alertness, inspired, status)
            risk = priest(sex, age, breath, o2sat, heart, systolic, temp, alertness, inspired, status)
      #      print("Risk for patient {} is {}".format(pid, risk))
            hospital_name = find_hospital(age, sex, risk)
            hospital_addr = get_address(hospital_name)
            resultline = line[1:11]
            resultline[1] = int(resultline[1])
            resultline[2] = int(resultline[2])            
            resultline[3] = float(resultline[3])
            resultline[4] = int(resultline[4])
            resultline[5] = int(resultline[5])            
            resultline[6] = float(resultline[6])            
            resultline.append(risk)
            resultline.append(hospital_name)
            resultline.append(hospital_addr)
            #print(resultline)
            results[pid] = resultline

    #print(json.dumps(results, indent=4))
    with open("my_people_results.json", "w") as outfile:
        json.dump(results, outfile)
        
    return results

````
